 ## model

```python
x=df.drop(columns='selling_price')
y=df['selling_price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
ohe=OneHotEncoder()
ohe.fit(x[['brand','fuel','owner']])
```

```python
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe.categories_
```

```python
col_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['brand','fuel','owner']),
                                 remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(col_trans)
```

```python
pipe=make_pipeline(col_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
sc=[]
for i in range(1500):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(col_trans,lr)
    pipe.fit(x_train,y_train)
    y_pred=pipe.predict(x_test)
    sc.append(r2_score(y_test,y_pred))
```

```python
np.argmax(sc)
```

```python
sc[np.argmax(sc)]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=np.argmax(sc))
lr=LinearRegression()
pipe=make_pipeline(col_trans,lr)
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
r2_score(y_test,y_pred)
```

```python
import pickle
```

```python
pickle.dump(pipe,open('LRModel.pkl','wb'))
```
